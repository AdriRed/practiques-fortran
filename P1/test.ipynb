```python
import math

def pk(k):
    return 3/5 * k**2 + math.e + 10*k

def s_m_n(m, n):
    a = 0
    for i in range(m, n+1):
        a = a + pk(i)
    return a

def s_asim(n):
    return (n**3)/5

s_m_n(8, 331)
```

```python
import matplotlib.pyplot as plt
import numpy as np

xs = [x for x in range(11, 312)]
ys = [s_m_n(8, x) for x in xs]
ys1 = [s_asim(x) for x in xs]
plt.scatter(xs, ys)
plt.plot(xs, ys1)

```
